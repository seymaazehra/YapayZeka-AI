```python
#kütüphaneleri import işlemi
from sklearn import datasets
iris=datasets.load_iris()
```

```python
#özelliklerin gösterimi
iris.get("feature_names")
```

```python
#özellik verilerin gösterme işlemi
iris.data
```

```python
#graf çizme işlemi
%matplotlib notebook
```

```python
#pca import işlemi
from sklearn.decomposition import PCA
```

```python
#boyut belirleme
pca=PCA(n_components=2)
```

```python
#train işlemi
pca.fit(iris.data) #öğretiyor
```

```python
X=pca.transform(iris.data) #boyutu ceviriyor
```

```python
import matplotlib.pyplot as plt
```

```python
#çizme fonksiyonu
plt.scatter(X[:,0], X[:,1], c=iris.target)
plt.scatter(X[10][0], [X[10][1]], s=[200], c=["r"], alpha=0.5)
```
