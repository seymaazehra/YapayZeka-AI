```python
import pandas as pd
from sklearn import linear_model
```

```python
#veri oku
iris =pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
names = ['sepal_length', 'sepal_width', 'petal_length', 'pedal_width', 'species'])

setosa=iris[iris.species=="Iris-setosa"]
X=setosa['sepal_length'].values.reshape(-1,1)
y=setosa['sepal_width'].values.reshape(-1,1)

#regresyon çalıştırır
ols=linear_model.LinearRegression()
model=ols.fit(X,y)

#check coefficients and evauate model
print("slope: {}\nintercept:{}".format(model.coef_[0][0], model.intercept_[0]))
print("r2: {}".format(model.score(X.y)))
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set_style("whitegrid")
sns.regplot(x="sepal_length", y="sepal_width", data=iris[iris.species=="Iris-setosa"])
plt.show()
```
